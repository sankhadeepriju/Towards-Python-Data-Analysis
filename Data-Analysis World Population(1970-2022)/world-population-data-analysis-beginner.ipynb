### Loading the dataset:

```python
population_df = pd.read_csv('world-population-dataset/world_population.csv')
population_df
```

```python
population_df.info()
```

### Renaming Columns to remove special characters and make it easy to access.

```python
population_df.rename(columns = {'Density (per km²)':'Density (per sq. km)'}, inplace = True)
population_df.rename(columns = {'Area (km²)':'Area (sq. km)'}, inplace = True)
population_df.rename(columns = {'Country/Territory':'Country'}, inplace = True)
population_df
```

```python
sorted_population = population_df.sort_values(by = 'Rank').reset_index(drop=True)
```

#### For ease of visualisation, let us check the population rise in South America.<br />
South America has the least number of countries in this dataset.

```python
sa_df = population_df[population_df.Continent == 'South America'].sort_values(by = 'Rank').reset_index(drop = True)
sa_df
```

```python
top_6_sa = sa_df.head(6)
top_6_sa
```

```python
x = ['1970', '1980', '1990', '2000', '2010', '2015', '2020', '2022']
y_bra = list(reversed(list(top_6_sa.loc[[0],'2022 Population':'1970 Population'].values[0])))
y_col = list(reversed(list(top_6_sa.loc[[1],'2022 Population':'1970 Population'].values[0])))
y_arg = list(reversed(list(top_6_sa.loc[[2],'2022 Population':'1970 Population'].values[0])))
y_per = list(reversed(list(top_6_sa.loc[[3],'2022 Population':'1970 Population'].values[0])))
y_ven = list(reversed(list(top_6_sa.loc[[4],'2022 Population':'1970 Population'].values[0])))
y_chl = list(reversed(list(top_6_sa.loc[[5],'2022 Population':'1970 Population'].values[0])))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(19, 10))
# Brazil
axes[0,0].plot(x, y_bra, 'o-y')
axes[0,0].set_xlabel('Years')
axes[0,0].grid(True)
axes[0,0].set_ylabel('Population (in 10000000)')
#axes[0,0].axes(facecolor = 'yellow')
axes[0,0].set_title('Population change in Brazil[1970-2022]');

# Colombia
axes[0,1].plot(x, y_col, 'd:b')
axes[0,1].set_xlabel('Years')
axes[0,1].grid(True)
axes[0,1].set_ylabel('Population (in 10000000)')
axes[0,1].set_title('Population change in Colombia[1970-2022]');

# Argentina 
axes[0,2].plot(x, y_arg, 'D-g')
axes[0,2].set_xlabel('Years')
axes[0,2].grid(True)
axes[0,2].set_ylabel('Population (in 10000000)')
axes[0,2].set_title('Population change in Argentina[1970-2022]');

# Peru 
axes[1,0].plot(x, y_per, 'p-.m')
axes[1,0].set_xlabel('Years')
axes[1,0].grid(True)
axes[1,0].set_ylabel('Population (in 10000000)')
axes[1,0].set_title('Population change in Peru[1970-2022]');

# Venezuela 
axes[1,1].plot(x, y_ven, 'x-c')
axes[1,1].set_xlabel('Years')
axes[1,1].grid(True)
axes[1,1].set_ylabel('Population (in 10000000)')
axes[1,1].set_title('Population change in Venezuela[1970-2022]');

# Brazil
axes[1,2].plot(x, y_chl, '*--r')
axes[1,2].set_xlabel('Years')
axes[1,2].grid(True)
axes[1,2].set_ylabel('Population (in 10000000)')
axes[1,2].set_title('Population change in Chile[1970-2022]');
```

We see that Venezuela had a drastic change in population after 2015. It probably means there were some event occurring that affected the Venezuela's people.<br /> Check the activities that might have lead to this population decrease for your General Knowledge.

### Let us check the change in population over 50 years \[1970-2020\]

```python
plt.figure(figsize = (30,10))
plt.grid(True)
X = list(asia_u25.Country)

Y1970 = asia_u25['1970 Population']
Y2020 = asia_u25['2020 Population']
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y1970, 0.4, label = '1970')
plt.bar(X_axis + 0.2, Y2020, 0.4, label = '2020')
  
plt.xticks(X_axis, X)
plt.xlabel('Country Code')
plt.ylabel('Population count in 100 Crores')
plt.title("Asian Countries(1-25) - Population over 50 years [1970-2020]")
plt.legend()
plt.show();
```

**This Bar Chart here shows you side by side population count in 1970 and how it changed in 2020.** <br /> We can clearly see that almost every Country had a significant increase (almost 2x) in population by 2020, as compared to population in 1970.

```python
plt.figure(figsize = (30,10))
plt.grid(True)
X = list(asia_l25.Country)

Y_1970 = asia_l25['1970 Population']
Y_2020 = asia_l25['2020 Population']
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y_1970, 0.4, label = '1970')
plt.bar(X_axis + 0.2, Y_2020, 0.4, label = '2020')
  
plt.xticks(X_axis, X)
plt.xlabel('Country Code')
plt.ylabel('Population count in 100 Crores')
plt.title("Asian Countries(26-50) - Population over 50 years [1970-2020]")
plt.legend()
plt.show();
```

This Bar Chart is pretty different from the countries we saw in our previous Chart.<br /> Here, we see Countries like Cambodia, Jordan, Kuwait, UAE and many more to have a drastic increase in population, whereas, in Countries like Hong Kong, Armenia and Georgia there isn't a huge change in population compared to other Asian Countries.

**Georgia :--**
Georgia, in fact, has a decrease in population over 50 years. There must have been some political/geographical factors responsible for such a impact that is pretty different from all other Asian Countries.

Let's look at Georgia's population change and try to know about more, so that we can trace back our study to understand the political/geographical causes that led to this situation.

```python
x_geo = ['1970', '1980', '1990', '2000', '2010', '2020']
y_geo = list(reversed(list(asia_df.loc[[40],'2010 Population':'1970 Population'].values[0])))
y_geo.append(asia_df.at[40,"2020 Population"])
plt.figure(figsize = (10,6))
plt.plot(x_geo, y_geo, '*-.c')
plt.xlabel('Years')
plt.grid(True)
plt.ylabel('Population')
plt.title('Population change in Georgia[1970-2020]');
```

Here, we see that Georgia had a rise in population till 1990 and it drastically fell within the next decade.<br /> Try finding the reason which factor led to the decline in population.

### Let's study the population change in UAE.

United Arab Emirates has an incredible rise in population. Let's check with the help of a line chart.

```python
x_uae = ['1970', '1980', '1990', '2000', '2010', '2020']
y_uae = list(reversed(list(asia_df.loc[[29],'2010 Population':'1970 Population'].values[0])))
y_uae.append(asia_df.at[29,"2020 Population"])
plt.figure(figsize = (9,5))
plt.plot(x_uae, y_uae, '*-.c')
plt.xlabel('Years')
plt.grid(True)
plt.ylabel('Population')
plt.title('Population change in United Arab Emirates[1970-2020]');
```

UAE has an incredible rise between 2000 - 2010. There must have been some social, industrial or political development that might have supported this huge population change. By the end of 2020, the population was seen to be almost 31x compared to the population of 1970.
